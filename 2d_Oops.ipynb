     lets there are two instances v1 and v2 what if we add them

     v1+v2 . but by default it does not happens. How ever we can do it manually by dunder....




```python
class Vector3D:

    def __init__(self,x1,x2,x3):

        self.x1=x1
        self.x2=x3
        self.x3=x3
    def __add__(self,other):
        if isinstance(other,Vector3D):
          return Vector3D(self.x1+other.x1, self.x2 + other.x2 , self.x3 + other.x3)
        else:
          raise ValueError("Addition only supported with other Vector")

     # representation dunder (Without it we only get obj at certain memory)

    def __repr__(self):
         return f"({self.x1}, { self.x2} , {self.x3})"

v1=Vector3D(10,20,30)
v2=Vector3D(1,4,6)

print(v1 + v2)
```

# Inheritance

it means there will be a parent class and child classes which will inherit the attributes, properties/methods from parent class and can have some extended methods/functions in child class.

```python
class Person:
  def __init__(self,name,age,job):
    self.name=name
    self.age=age
    self.job=job

  def print_name(self):
    print(f'Name is {self.name}!')

  def __repr__(self):
    return f"name: {self.name},Age : {self.age},Job: {self.job}"

class Programmer(Person):
    def __init__(self,name,age,job,main_language):
      super().__init__(name,age,job)
      self.main_language =  main_language

    def __repr__(self): # overriding Represenation dunder
      return f"Name: {self.name}, Age: { self.age} , Job:{self.job} , Language : {self.main_language}"

prog1=Programmer('Mike',30, 'Programmer','C++')
prog1.print_name()
print(prog1.main_language)
print(prog1)
```

```python
class Flyer:
  def fly(self):
    return "I can fly!"

class Swimmer:
  def swim(self):
    return "I can swim!"

class FlyingFish(Flyer,Swimmer):
  pass

ff=FlyingFish()
print(ff.fly())
print(ff.swim())
```

# Polymorphism

```python
import math

class Shape:
  def area(self):
    raise NotImplementedError

class Circle(Shape):
    def __init__(self,radius):
      self.radius = radius

    def area(self):
      return math.pi * self.radius**2
class Square(Shape):

  def __init__(self,side):
     self.side = side

  def area(self):
     return self.side * self.side

class Triangle(Shape):
    def __init__(self,base,height):
       self.base=base
       self.height=height

    def area(self):
      return 0.5 * self.base * self.height
if __name__ ==  "__main__":
      shapes=[Circle(2), Square(3), Triangle(4,5)]
      areas=[x.area() for x in shapes]
      print(areas)
```
